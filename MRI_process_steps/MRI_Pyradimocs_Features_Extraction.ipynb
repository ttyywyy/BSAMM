```python
import radiomics
from radiomics import featureextractor
import os
import pandas as pd
import SimpleITK as sitk
```

```python
settings = {}
settings['binWidth'] = 30
settings['resampledPixelSpacing'] = [0.5,0.5,0.5] # unit: mm
settings['interpolator'] = sitk.sitkNearestNeighbor
settings['normalize'] = True
settings['geometryTolerance'] = 1e-16
```

```python
path_Tr = 'xxx/images'
path_Lr = 'xxx/labels'
```

```python
basePath = 'XXX/step2'
```

```python
l = os.listdir(path_Tr)

df = pd.DataFrame()
extractor = featureextractor.RadiomicsFeatureExtractor(**settings)
extractor.enableAllFeatures()
extractor.enableImageTypes(Original={}, LoG={"sigma" : [2.0,3.0,4.0,5.0]}, Wavelet={})
for i in l:
    if i.endswith('.gz'):
        if not i.startswith('.'):
            img = os.path.join(path_Tr,i)
            mask = os.path.join(path_Lr,i)
            print(i)
            try: 
                featureVector = extractor.execute(img, mask)
                df_new = pd.DataFrame.from_dict(featureVector.values()).T
                df_new.columns = featureVector.keys()
                df_new['id'] = i[:-7]
                df = pd.concat([df,df_new])
            except:
                pass
df.to_excel(os.path.join(basePath,'XXX_radiomics.xlsx'))
```
